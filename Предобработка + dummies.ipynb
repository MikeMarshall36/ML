```python
import pandas as pd
import numpy  as np
import matplotlib
import seaborn as sns
import matplotlib.pyplot as plt
```

0. Загрузка датасета

```python
house = pd.read_csv('House_Rent_Dataset.csv')
house
```

1. Признак floor делим на total_floors и floor_number

```python
floor_num_array = []
floors_array = []

for i in range(house.shape[0]):
  floor_num_array.append(str(house['Floor'][i]).replace('\n', ' ').split(' ')[0])
  floors_array.append(str(house['Floor'][i]).replace('\n', ' ').split(' ')[-1])
floor_num_array = pd.Series(floor_num_array)
floors_array = pd.Series(floors_array)
```

```python
house['Floor_number'] = floor_num_array
house['Total_floors'] = floors_array
house = house.drop('Floor', axis=1)
house['Floor_number'] = house['Floor_number'].replace('Ground', 0)
house['Floor_number'] = house['Floor_number'].replace('Upper', 0)
house['Floor_number'] = house['Floor_number'].replace('Lower', 0)
house
```

2. Работаем с признаками Area Type и Point of Contact

```python
print(house['Area Type'].unique())
print(house['Point of Contact'].unique())

print(house['Point of Contact'].value_counts())
print(house['Area Type'].value_counts().index[-1])

house['Point of Contact'] = house['Point of Contact'].replace(house['Point of Contact'].value_counts().index[-1],  house['Point of Contact'].mode()[0])
house['Area Type'] = house['Area Type'].replace(house['Area Type'].value_counts().index[-1],  house['Area Type'].mode()[0])

print(house['Area Type'].unique())
print(house['Point of Contact'].unique())


house
```

```python
house['Area Type'] = house['Area Type'].replace('Super Area', 0)
house['Area Type'] = house['Area Type'].replace('Carpet Area', 1)

house['Point of Contact'] = house['Point of Contact'].replace('Contact Owner', 0)
house['Point of Contact'] = house['Point of Contact'].replace('Contact Agent', 1)

house
```

Работа с City

```python
cities = house['City'].unique()
print(len(cities), cities)

columns = {}
for pos, city in enumerate(cities):
  columns[pos] = city

city_encoded = pd.get_dummies(house['City']).rename(columns=columns)
house = house.join(city_encoded).drop('City', axis=1)
house
```

4. Работа с Furnising Status

```python
print(house['Furnishing Status'].unique())
```

```python
house['Furnishing Status'] = house['Furnishing Status'].replace('Unfurnished', 0)
house['Furnishing Status'] = house['Furnishing Status'].replace('Semi-Furnished', 1)
house['Furnishing Status'] = house['Furnishing Status'].replace('Furnished', 2)
house
```

5. Рабоат с Tenant Preferred

```python
house = house.join(pd.get_dummies(house["Tenant Preferred"])).drop('Tenant Preferred', axis=1)
house["Bachelors"]+=house["Bachelors/Family"]
house["Family"]+=house["Bachelors/Family"]
house=house.drop("Bachelors/Family", axis=1)
house.head()
```

6. Работа с Posted On

```python
house['Posted On'] = pd.DatetimeIndex(house['Posted On']).month
print(house['Posted On'].value_counts())
```

```python
house.to_csv("new_House_Rent_Dataset.csv", encoding='utf-8', index=False)
house
```
